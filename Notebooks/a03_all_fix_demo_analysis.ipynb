```python
fix_demo_arr = []
for lang in lang_list:
    fix_demo_arr.append(pd.read_csv(f'../Datasets/DataToUse/{lang}_fix_demo.csv'))

all_fix_demo = pd.concat(fix_demo_arr)
```

```python
l2_cols = ['L2_spelling_skill', 'L2_vocabulary_size', 'L2_lexical_skill']
fix_cols = ['Fix_X', 'Fix_Y', 'Fix_Duration']
```

```python
l2_corr = all_fix_demo[l2_cols].corr()

fig = plt.figure(figsize=(6, 6))
sns.heatmap(
    data=l2_corr,
    xticklabels=l2_corr.columns,
    yticklabels=l2_corr.columns,
    cmap="YlGnBu",
    annot=True
)
plt.show()
```

```python
l2_fix_corr = all_fix_demo[l2_cols + fix_cols].corr()

fig = plt.figure(figsize=(8, 8))
sns.heatmap(
    data=l2_fix_corr,
    xticklabels=l2_fix_corr.columns,
    yticklabels=l2_fix_corr.columns,
    cmap="YlGnBu",
    annot=True
)
plt.show()
```
